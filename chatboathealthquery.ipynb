```python
from groq import Groq
 
GROQ_API_KEY = "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"  # 👈 paste your key here
 
client = Groq(api_key=GROQ_API_KEY)
 
SYSTEM_PROMPT = """
You are a helpful and friendly medical information assistant.
Rules:
- Answer general health questions.
- Use simple language.
- Do not diagnose diseases.
- Do not prescribe medicines.
- Recommend a doctor for serious symptoms.
- Keep answers short and clear.
- End with: Please consult a healthcare professional for advice.
"""
 
BLOCKED_KEYWORDS = [
    "overdose",
    "self harm",
    "suicide",
    "illegal drug"
]
 
conversation_history = []
 
def health_chatbot(question):
    # Safety filter
    for word in BLOCKED_KEYWORDS:
        if word in question.lower():
            return "I cannot assist with this query. Please contact a healthcare professional."
 
    # Add user message to history
    conversation_history.append({
        "role": "user",
        "content": question
    })
 
    try:
        response = client.chat.completions.create(
          model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT}
            ] + conversation_history,
            max_tokens=200,
            temperature=0.4,
        )
 
        bot_reply = response.choices[0].message.content.strip()
 
        # Save to history
        conversation_history.append({
            "role": "assistant",
            "content": bot_reply
        })
 
        return bot_reply
 
    except Exception as e:
        return "Error: " + str(e)
 
print("✅ Chatbot is ready!")
```

```python
print('Health Query Chatbot — Type quit to exit')
print('=' * 50)
 
while True:
    user_input = input('You: ').strip()
 
    if user_input.lower() in ['quit', 'exit', 'bye']:
        print('Goodbye! Stay healthy!')
        break
 
    if not user_input:
        continue
 
    response = health_chatbot(user_input)
    print(f'Bot: {response}')
    print()
```
